```python
import pandas as pd

# Load the Hadith dataset
hadith_df = pd.read_csv("hadith_dataset.csv")

# Define keywords for each pillar
pillar_keywords = {
    "Shahada": ["faith", "testimony", "belief"],
    "Salat": ["prayer", "worship", "daily prayer"],
    "Zakat": ["charity", "almsgiving", "giving"],
    "Sawm": ["fasting", "Ramadan", "abstaining"],
    "Hajj": ["pilgrimage", "Mecca", "Kaaba"]
}

# Function to check if a Hadith contains any of the keywords
def contains_keywords(hadith_text, keywords):
    for keyword in keywords:
        if keyword.lower() in hadith_text.lower():
            return True
    return False

# Extract Hadith related to the Five Pillars
hadith_results = {}
for pillar, keywords in pillar_keywords.items():
    pillar_hadith = hadith_df[hadith_df['text_en'].apply(contains_keywords, keywords=keywords)]['text_en'].tolist()
    hadith_results[pillar] = pillar_hadith

# Print the extracted Hadith
for pillar, hadiths in hadith_results.items():
    print(f"Pillar: {pillar}")
    for hadith in hadiths:
        print(f"Hadith: {hadith}")
    print("\n")
```

```python
import pandas as pd

# Load the Quran dataset
quran_df = pd.read_csv("Quran_English.csv")

# Define keywords for each pillar
pillar_keywords = {
    "Shahada": ["faith", "testimony", "belief"],
    "Salat": ["prayer", "worship", "daily prayer"],
    "Zakat": ["charity", "almsgiving", "giving"],
    "Sawm": ["fasting", "Ramadan", "abstaining"],
    "Hajj": ["pilgrimage", "Mecca", "Kaaba"]
}

# Function to check if an Ayah contains any of the keywords
def contains_keywords(verse_text, keywords):
    for keyword in keywords:
        if keyword.lower() in verse_text.lower():
            return True
    return False

# Extract Quran Ayahs related to the Five Pillars
quran_results = {}
for pillar, keywords in pillar_keywords.items():
    pillar_ayahs = quran_df[quran_df['Verse'].apply(contains_keywords, keywords=keywords)]['Verse'].tolist()
    quran_results[pillar] = pillar_ayahs

# Print the extracted Ayahs
for pillar, ayahs in quran_results.items():
    print(f"Pillar: {pillar}")
    for ayah in ayahs:
        print(f"Ayah: {ayah}")
    print("\n")
```
