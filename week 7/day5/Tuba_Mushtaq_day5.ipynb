```python
import pandas as pd

# Load the Hadith dataset
hadith_df = pd.read_csv("hadith_dataset.csv")

# Define keywords for each pillar
pillar_keywords = {
    "Shahada": ["faith", "testimony", "belief"],
    "Salat": ["prayer", "worship", "daily prayer"],
    "Zakat": ["charity", "almsgiving", "giving"],
    "Sawm": ["fasting", "Ramadan", "abstaining"],
    "Hajj": ["pilgrimage", "Mecca", "Kaaba"]
}

# Function to check if a Hadith contains any of the keywords
def contains_keywords(hadith_text, keywords):
    for keyword in keywords:
        if keyword.lower() in hadith_text.lower():
            return True
    return False

# Get user input to select a pillar
while True:
    user_input = input("Enter one of the Five Pillars (Shahada, Salat, Zakat, Sawm, Hajj): ").strip().capitalize()
    if user_input in pillar_keywords:
        break
    else:
        print("Invalid input. Please enter a valid pillar.")

# Extract Hadith related to the selected pillar
selected_keywords = pillar_keywords[user_input]
selected_hadiths = hadith_df[hadith_df['text_en'].apply(contains_keywords, keywords=selected_keywords)]['text_en'].tolist()

# Print the extracted Hadiths
print(f"Related Hadiths for {user_input}:\n")
for hadith in selected_hadiths:
    print(hadith)
```
