```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv('/kaggle/input/weight-and-heightcsv/weight-height.csv')
data.head()
```

```python
data.describe()
```

```python
data
```

# Mean 

```python
height_mean = np.mean(data['Height'])
weight_mean = np.mean(data['Weight'])
print("Mean of Height: ",height_mean)
print("Mean of weight: ",weight_mean)
```

# Median

```python
height_med = np.median(data['Height'])
weight_med = np.median(data['Weight'])
print("Median of Height: ",height_med)
print("Median of weight: ",weight_med)
```

# Standard Deviation

```python
height_std = np.std(data['Height'])
weight_std = np.std(data['Weight'])
print("Standard Deviation of Height: ",height_std)
print("Standard Deviation of weight: ",weight_std)
```

# skewness and kurtosis

```python
h = data['Height']
w = data['Weight']
```

```python
print('Height skewness', stats.skew(h))
```

```python
print('Height kurtosis', stats.kurtosis(h))
```

```python
print('Weight skewness', stats.skew(w))
```

```python
print('Weight kurtosis', stats.kurtosis(w))
```
